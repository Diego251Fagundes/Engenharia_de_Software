  19. Crie dois arrays de números inteiros aleatórios de tamanho 20 e calcule o
ângulo entre eles (produto escalar normalizado).

```python
import numpy as np

array_1 = np.random.randint(1,10,20)
array_2 = np.random.randint(1,10,20)
print(f"Array 1: \n{array_1}")
print(f"\nArray 2: \n{array_2}")

# Calcular o produto escalar
produto_escalar = np.dot(array_1, array_2)

# Calcular a norma de cada array
norma_1 = np.linalg.norm(array_1)
norma_2 = np.linalg.norm(array_2)

# Aplica a fórmula com os valores encontrados.
cosseno_angulo = produto_escalar / (norma_1 * norma_2)

# usa o np.arccos para descobrir o angulo o resultado será em radianod
angulo_em_radianos = np.arccos(cosseno_angulo)

# Converte o resultado para Graus
angulo_em_graus = np.degrees(angulo_em_radianos)

print(f"\nNorma do Array 1: {norma_1:.2f}")
print(f"Norma do Array 2: {norma_2:.2f}")
print(f"Cosseno do Ângulo: {cosseno_angulo:.4f}")
print(f"\nÂngulo em Radianos: {angulo_em_radianos:.4f}")
print(f"Ângulo em Graus: {angulo_em_graus:.2f}°")
```

  10. Calcule a integral dupla da função f(x, y) = xy + x² no intervalo [0, 2]
para x e [0, 1] para y.

```python
import numpy as np
from scipy.integrate import dblquad

# Definir a função: a função deve ser no formato f(y, x). A variável da integral de dentro (y) vem primeiro.
def f(y, x):
    """Função f(x, y) = xy + x²"""
    return x * y + x**2

# Limites para a variável de FORA (x)
x_min = 0
x_max = 2

# Limites para a variável de DENTRO (y)
y_min = lambda x: 0
y_max = lambda x: 1

# Calcula a integral dupla
resultado, erro = dblquad(f, x_min, x_max, y_min, y_max)

print(f"O resultado da integral dupla é: {resultado:.4f}")
print(f"A estimativa do erro absoluto é: {erro:.2e}") # .2e formata em notação científica
```

  15. Calcule a transformada de Fourier inversa de um sinal complexo gerado
aleatoriamente.

```python
import numpy as np
from scipy.fft import ifft

# Gera um Sinal Complexo Aleatório no Domínio da Frequência
N = 10

# Um número complexo tem a forma a + bi.
# Gera as partes real (a) e imaginária (b) com números aleatórios.
parte_real = np.random.randn(N)
parte_imaginaria = np.random.randn(N)

# Combina as duas partes usando '1j' (a unidade imaginária em Python)
sinal_frequencia = parte_real + 1j * parte_imaginaria

np.set_printoptions(precision=4, suppress=True)
print("Sinal de Frequência (Complexo e Aleatório):")
print(sinal_frequencia)
print("\n")


# Calcula a Transformada de Fourier Inversa (IFFT)
# A função ifft reconstrói o sinal no domínio do tempo a partir das frequências.
sinal_tempo = ifft(sinal_frequencia)

# O resultado da IFFT também é um array de números complexos.
print("Sinal no Domínio do Tempo (resultado da IFFT):")
print(sinal_tempo)
```

  16. Resolva a equação diferencial de segunda ordem (d²y/dx²) + y = 0, com condições iniciais y(0) = 1 e y′(0) = 0.

```python
import numpy as np
from scipy.integrate import solve_ivp

# Definir a Função do Sistema de EDOs de Primeira Ordem
# A EDO y'' = -y é convertida para o sistema S' = [y', z'] = [z, -y],
# onde o vetor de estado é S = [y, z] e z = y'.
def sistema(x, S):
    y, z = S
    return [z, -y]

# Condições iniciais
S0 = [1.0, 0.0]

# Intervalo no qual queremos resolver a EDO (de x=0 a x=4)
intervalo_x = [0, 4]
# Pontos onde queremos que o SciPy calcule a solução.
# Vamos pegar alguns pontos para poder comparar na tabela.
pontos_x_eval = np.linspace(intervalo_x[0], intervalo_x[1], 9)


# Resolve a EDO
solucao = solve_ivp(sistema, intervalo_x, S0, t_eval=pontos_x_eval)

# A solução para 'y' está na primeira linha do resultado solucao.y
y_resultado_scipy = solucao.y[0]
x_resultado = solucao.t

print("="*65)
print("Solução da EDO y'' + y = 0 com y(0)=1, y'(0)=0")
print("="*65)

# A solução analítica (exata) para este problema é y(x) = cos(x).
# compara o resultado numérico do SciPy com a solução exata
# em uma tabela para verificar a precisão.

print("\nComparando a Solução Numérica (SciPy) com a Solução Analítica (cos(x)):\n")
print(f"{'Valor de x':>12s} | {'y (SciPy)':>15s} | {'y (cos(x))':>15s} | {'Diferença':>15s}")
print("-" * 65)

# Loop para imprimir cada ponto calculado
for i in range(len(x_resultado)):
    x_val = x_resultado[i]
    y_scipy = y_resultado_scipy[i]
    y_analitico = np.cos(x_val)
    diferenca = np.abs(y_scipy - y_analitico)

    print(f"{x_val:12.4f} | {y_scipy:15.8f} | {y_analitico:15.8f} | {diferenca:15.2e}")
```

  19. Realize a transformada de Fourier de um sinal discreto composto de senos
e cossenos.

```python
import numpy as np
from scipy.fft import fft, fftfreq

# 1. Definição de um sinal discreto composto de senos e cossenos
N = 200  # Número de amostras do sinal
TAXA_AMOSTRAGEM = 100 # Amostras por segundo

t = np.linspace(0.0, 2.0, N, endpoint=False)
sinal_discreto = np.sin(2 * np.pi * 10 * t) + 0.5 * np.cos(2 * np.pi * 20 * t)

# 2. Cálculo da Transformada de Fourier
# A função fft calcula as amplitudes complexas do sinal.
transformada_fft = fft(sinal_discreto)

# A função fftfreq calcula as frequências correspondentes a cada ponto da transformada.
frequencias_hz = fftfreq(N, 1 / TAXA_AMOSTRAGEM)

# 3. Apresentação dos resultados brutos
print("--- Resultado da Transformada de Fourier ---")

print("\nAmplitudes Complexas (Saída da FFT):")
print(transformada_fft)

print("\nFrequências Correspondentes (Saída da FFTFREQ em Hz):")
print(frequencias_hz)
```

  20. Resolva um sistema de equações lineares com três variáveis utilizando
SciPy.

```python
import numpy as np
from scipy.linalg import solve

# --- Passo 1: Montar as Matrizes do Sistema ---
# A matriz A contém os coeficientes de x, y, z em cada equação.
# Cada linha da matriz corresponde a uma equação.
A = np.array([
    [1, 1, 1],   # Coeficientes da Equação 1
    [2, -1, 1],  # Coeficientes da Equação 2
    [-1, 2, 2]   # Coeficientes da Equação 3
])

# O vetor b contém os resultados (termos constantes) de cada equação.
b = np.array([4, 8, 2])

# --- Passo 2: Resolver o Sistema com SciPy ---
# A função linalg.solve(A, b) funciona perfeitamente para qualquer
# tamanho de sistema, desde que a matriz A seja quadrada.
solucao = solve(A, b)


# --- Passo 3: Apresentar a Solução ---
print("Solução Encontrada:")
# A solução é um array com os valores de x, y, e z, nessa ordem.
print(f"O vetor solução [x, y, z] é: {solucao}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar os dados da função principal
# Usar múltiplos de PI para o seno deixa o gráfico mais completo e fácil de interpretar.
x = np.linspace(0, 4 * np.pi, 200)
y = np.sin(x)

# Simular os limites do intervalo de confiança
# Definir uma margem de erro fixa de 0.3 para este exemplo.
erro = 0.3
limite_superior = y + erro
limite_inferior = y - erro

# Plota a linha principal
plt.plot(x, y, label='y = sin(x)')

# Esta é a função principal para criar a área sombreada(Intervalo de confiança).
# O parâmetro 'alpha' controla a transparência, o que é essencial para o efeito.
plt.fill_between(x, limite_inferior, limite_superior, color='blue', alpha=0.2, label='Intervalo de Confiança')
plt.title("Seno com Intervalo de Confiança")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```

  16. Crie um gráfico de linha com títulos para cada eixo e adicione um rótulo
de texto no ponto x = 5.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
x = np.linspace(0, 10, 100)
y = (x - 5)**2

# Criar o gráfico de linha e adicionar títulos aos eixos
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = (x-5)²')
plt.title('Gráfico com Rótulo de Texto em um Ponto Específico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid(True, linestyle='--', alpha=0.5)

# Adicionar o rótulo de texto no ponto x=5 definindo as cordenadas
x_ponto = 5
y_ponto = (x_ponto - 5)**2 # y = 0

# A função plt.text() recebe a coordenada x, a coordenada y e o texto.
plt.text(x_ponto, y_ponto, 'Ponto de Mínimo (x=5)', fontsize=12, color='red')

plt.legend()
plt.show()
```

  18. Crie um gráfico de barras com valores positivos e negativos e use diferentes cores para cada tipo de valor.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']
resultados = np.array([30, -12, 50, -22, 45]) # Lucros (positivo) e perdas (negativo)

# Passo 2: Criar a lista de cores condicionalmente
cores = []
for valor in resultados:
    if valor >= 0:
        cores.append('g') # Verde para valores positivos (lucro)
    else:
        cores.append('r') # Vermelho para valores negativos (perda)

# Criar o gráfico de barras passando a lista de cores
plt.bar(meses, resultados, color=cores)

# Criar uma linha no 0
plt.axhline(0, color='grey', linewidth=2)

# Titulos e rótulos
plt.title('Resultado Financeiro Mensal (Lucro/Perda)', fontsize=16)
plt.ylabel('Valor (em milhares de R$)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```

  19. Plote um gráfico de linha com duas séries e adicione uma legenda para
diferenciá-las.

```python
import matplotlib.pyplot as plt
import numpy as np

# Preparar os dados para as duas séries
meses = np.arange(1, 13)
# Dados de vendas simulados para dois produtos
vendas_produto_A = 50 + 15 * np.sin(meses * 0.8) + np.random.randn(12) * 2
vendas_produto_B = 70 + 10 * np.cos(meses * 0.7) + np.random.randn(12) * 2

# Plotar as duas séries de dados
plt.figure(figsize=(10, 6))

# Série 1: Produto A, linha tracejada azul com marcadores circulares
plt.plot(meses, vendas_produto_A,
         color='blue',
         linestyle='--',
         marker='o',
         label='Produto A') #Definir o rótulo aqui

# Série 2: Produto B, linha sólida vermelha com marcadores 'x'
plt.plot(meses, vendas_produto_B,
         color='red',
         linestyle='-',
         marker='x',
         label='Produto B') # Definir o rótulo aqui

# Titulo e rotulos
plt.title('Comparativo de Vendas Mensais', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Unidades Vendidas', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.xticks(meses) # Garante que todos os meses (1 a 12) apareçam no eixo X
plt.legend()

plt.show()
```


20. Adicione uma grade com espaçaamento personalizado a um gráfico de linha para y = x².

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
x = np.linspace(0, 10, 100)
y = x**2

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = x²')

# Para o eixo X, um tick a cada 2 unidades (0, 2, 4, ..., 10)
plt.xticks(np.arange(0, 11, 2))
# Para o eixo Y, um tick a cada 10 unidades (0, 10, 20, ..., 100)
plt.yticks(np.arange(0, 101, 10))

plt.grid(True, linestyle='--', alpha=0.7)

# Titulo e rotulos
plt.title('Gráfico com Espaçamento de Grade Personalizado', fontsize=16)
plt.xlabel('Eixo X', fontsize=12)
plt.ylabel('Eixo Y', fontsize=12)
plt.legend()

plt.show()
```
